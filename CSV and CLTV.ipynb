```python
txid_to_spend = "c66fc1128848dcfcbf2261c0c744c26c57a807adfc912b23375a1a553bae1400"
vout_to_spend = 0
total_amt = 625 * 10 ** 6
total_amt
input_amt = total_amt // 2
input_amt
```

```python
from codes import size_compact_size, cleanup_tx
from hashlib import sha256


cltv_wsh = sha256(bytes.fromhex(cltv_hex)).hexdigest()
csv_wsh = sha256(bytes.fromhex(csv_hex)).hexdigest()

lockup_tx = cleanup_tx(
f"""
version: 0200 0000
inputs: 01
    txid: {bytes.fromhex(txid_to_spend)[::-1].hex()}
    vout: {vout_to_spend.to_bytes(4, 'little')}
    scriptSig: 00
    sequence: ffff ffff
outputs: 02
    amount: {(input_amt - 400).to_bytes(8, 'little').hex()}
    scriptPubKey: 22 00 20 {cltv_wsh}
    amount: {(input_amt - 400).to_bytes(8, 'little').hex()}
    scriptPubKey: 22 00 20 {csv_wsh}
locktime:
    0000 0000
""")
lockup_tx
```

```python
signed_lockup_tx = !bitcoin-cli -regtest signrawtransactionwithwallet "$lockup_tx" | jq -r .hex
signed_lockup_tx = signed_lockup_tx[0]

!bitcoin-cli -regtest signrawtransactionwithwallet "$lockup_tx"
```

```python
!bitcoin-cli -regtest sendrawtransaction "$signed_lockup_tx"
sent_txid = sent_txid[0]
```

```python
sent_txid
```

```python
spend_csv_tx = cleanup_tx(
f"""
version: 0200 0000
marker+flag: 0001
inputs: 01
    txid: {bytes.fromhex(sent_txid)[::-1].hex()}
    vout: {(1).to_bytes(4, 'little')}
    scriptSig: 00
    sequence: {CSV.to_bytes(4, 'little').hex()}
outputs: 01
    amount: {(input_amt - 800).to_bytes(8, 'little').hex()}
    scriptPubKey: 22 00 20 {csv_wsh}
witness: 
01
    06 {csv_hex}
locktime:
    0000 0000
""")
           
spend_csv_tx
!bitcoin-cli -regtest decoderawtransaction "$spend_csv_tx"
```

```python
spend_cltv_tx = cleanup_tx(
f"""
version: 0200 0000
marker + flag: 0001
inputs: 01
    txid: {bytes.fromhex(sent_txid)[::-1].hex()}
    vout: {(0).to_bytes(4, 'little')}
    scriptSig: 00
    sequence: fdff ffff
outputs: 01
    amount: {(input_amt - 800).to_bytes(8, 'little').hex()}
    scriptPubKey: 22 00 20 {cltv_wsh}
witnesses:
    01
    06 {cltv_hex}
locktime:
    {CLTV.to_bytes(4, 'little').hex()}
""")

print(spend_cltv_tx)
!bitcoin-cli -regtest decoderawtransaction "$spend_cltv_tx"
```
